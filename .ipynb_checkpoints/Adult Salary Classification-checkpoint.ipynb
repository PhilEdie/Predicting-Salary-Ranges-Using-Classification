```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import sklearn
from scipy import stats
from warnings import filterwarnings
filterwarnings('ignore')
```

### Loading the CSV files and adding names to the columns.

#### Column names were sourced from: https://archive.ics.uci.edu/ml/datasets/adult

```python
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
                'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary']

adult = pd.read_csv('adult.data', names = column_names)
adult_test = pd.read_csv('adult.test', names = column_names)
adult_test = adult_test.drop(0).reset_index(drop = True)
```

### Changed train and test salary formats to be consistent. 

```python
print(adult['salary'].unique())
print(adult_test['salary'].unique())

#Rename Salary instances so they are consistent. 
adult_test['salary'] = adult_test['salary'].str.replace(' <=50K.', ' <=50K')
adult_test['salary'] = adult_test['salary'].str.replace(' >50K.', ' >50K')
```

### Changing datasets to use X_train, y_train format. I also encoded the salary column to a binary 0 or 1. 

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

X_train = adult.drop('salary', axis = 1)
X_test = adult_test.drop('salary', axis = 1)
y_train = enc.fit_transform(adult['salary'])
y_test = enc.transform(adult_test['salary'])



#Removing 'salary' from column names since it is what we're predicting.
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
                'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country']

```

### Replacing '?' with NaN:

```python
X_train.replace('[\?,)]', np.nan, regex=True, inplace = True)
X_test.replace('[\?,)]', np.nan, regex=True, inplace = True)
```

### Handle Missing values: 

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'most_frequent')
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns = column_names)
X_test = pd.DataFrame(imputer.transform(X_test), columns = column_names)
```

### Convert rare categories to category "Other"

####  With help from:
#### https://stackoverflow.com/questions/53648208/rename-the-less-frequent-categories-by-other-python
#### https://www.analyticsvidhya.com/blog/2021/05/how-to-perform-one-hot-encoding-for-multi-categorical-variables/

```python
for cat in categories:
    top_10 = [x for x in X_train[cat].value_counts().sort_values(ascending = False).head(10).index]
    X_train[cat] = np.where(X_train[cat].isin(top_10), X_train[cat], 'Other')
    X_test[cat] = np.where(X_test[cat].isin(top_10), X_test[cat], 'Other')
```

### Encoding categorical data using dummy variables:

```python
for cat in categories:

    X_train_encoded = pd.get_dummies(X_train[cat], drop_first = True)
    X_train = X_train.merge(X_train_encoded, left_index = True, right_index = True)
    X_train.drop(cat, axis = 1, inplace = True)
    
    X_test_encoded = pd.get_dummies(X_test[cat], drop_first = True)
    X_test = X_test.merge(X_test_encoded, left_index = True, right_index = True)
    X_test.drop(cat, axis = 1, inplace = True)
```

### Applying Scaling to data:

```python
from sklearn.preprocessing import StandardScaler

X_train_numeric = X_train[['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']]
X_test_numeric = X_test[['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']]

X_col_names = list(X_train_numeric.columns)

scaler=StandardScaler()

#Transform training data: -------------------

X_train_numeric = pd.DataFrame(scaler.fit_transform(X_train_numeric), columns = X_col_names)

# Transform testing data: ---------------------

X_test_numeric = pd.DataFrame(scaler.transform(X_test_numeric), columns = X_col_names)

display(X_train_numeric)
```

### Combining dummy variables with original dataset:

```python
X_train.drop(['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week'], axis = 1, inplace = True)
X_test.drop(['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week'], axis = 1, inplace = True)


X_train = pd.merge(X_train_numeric, X_train, left_index = True, right_index = True)
X_test = pd.merge(X_test_numeric, X_test, left_index = True, right_index = True)

display(X_train)
```

### Importing classifiers: 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
```

### Importing performance metrics: 

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
"""
    Fits and makes predictions using the input model.
    Calculates the performance metrics for the fitted model. 
    returns a dataframe row which includes performance metrics for accuracy, recall, f1, and auc. 
"""
def calculate_pms(clf, clf_name):
    print("Calculating pms for", clf_name)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
 
    acc = accuracy_score(y_test, y_pred)
    acc = round(acc, 2)

    recall = recall_score(y_test, y_pred)
    recall = round(recall, 2)

    f1 = f1_score(y_test, y_pred)
    f1 = round(f1, 2)

    y_pred_prob = clf.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:,1], pos_label=1)
    auc_score = auc(fpr, tpr)
    auc_score = round(auc_score, 2)

    row = {'Algorithm' : clf_name, 'Accuracy' : acc, 'Recall' : recall, 'F1':f1, 'AUC': auc_score}
    return row
    
```

### Create all the classifiers and calculate performance metrics:

```python
#Parameters for the below classifiers have been found using randomized grid searches. Please see the Hyperparameter
#Optimisation section. 

list_of_clfs = [KNeighborsClassifier( p = 2, n_neighbors = 18, leaf_size = 10), 
                GaussianNB(), 
                SVC(probability=True), 
                DecisionTreeClassifier(splitter = 'best', min_weight_fraction_leaf = 0.1, min_samples_leaf = 6, max_leaf_nodes = 80, max_features = None, max_depth = 9), 
                RandomForestClassifier(n_estimators= 100, min_samples_split = 15, min_samples_leaf = 1, max_depth = 25),
                AdaBoostClassifier(), GradientBoostingClassifier(), LinearDiscriminantAnalysis(), 
                MLPClassifier(solver = 'adam', learning_rate = 'adaptive', hidden_layer_sizes = (100,), alpha = 0.05, activation = 'tanh'),
                LogisticRegression()]

clf_names = ['KNeighborsClassifier', 'GaussianNB', 'SVC', 'DecisionTreeClassifier', 'RandomForestClassifier', 
             'AdaBoostClassifier', 'GradientBoostingClassifier', 'LinearDiscriminantAnalysis', 'MLPClassifier',
             'LogisticRegression']


pm = pd.DataFrame(columns = ['Algorithm', 'Accuracy', 'Recall', 'F1', 'AUC'])

for i in range(len(list_of_clfs)):
    pm = pm.append(calculate_pms(list_of_clfs[i], clf_names[i]), ignore_index = True)

display(pm)
```

### Calculate ranks for each of the classifiers: 

```python
pm_ranked = pm.rank(method = 'max', numeric_only = True, ascending = False).astype(int)
pm_ranked['Algorithm'] = pm['Algorithm']
pm_ranked = pm_ranked[['Algorithm', 'Accuracy', 'Recall', 'F1', 'AUC']]
pm_ranked['Sum Of Ranks'] = pm_ranked[list(pm_ranked.columns)].sum(axis=1)
pm_ranked['Overall Rank'] = pm_ranked['Sum Of Ranks'].rank(method = 'min').astype(int)
pm_ranked.sort_values(['Overall Rank'], inplace = True)
display(pm_ranked)
```
